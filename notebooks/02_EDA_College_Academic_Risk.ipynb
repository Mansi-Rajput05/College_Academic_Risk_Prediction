```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("dark_background")
sns.set_context("talk")
```

```python
df = pd.read_csv("../data/processed/student_featured.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="risk_category", data=df)
plt.title("Distribution of Academic Risk Categories")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df["final_score"], bins=20, kde=True)
plt.title("Distribution of Final Academic Score")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="risk_category", y="attendance_percentage", data=df)
plt.title("Attendance vs Academic Risk")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x="risk_category", y="internal_marks_2", data=df)
plt.title("Internal Marks vs Academic Risk")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x="backlog_count", hue="risk_category", data=df)
plt.title("Backlogs vs Risk Category")
plt.show()
```

```python
plt.figure(figsize=(10,6))
corr = df.select_dtypes(include=np.number).corr()

sns.heatmap(
    corr,
    cmap="coolwarm",
    annot=False,
    linewidths=0.5
)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

sns.boxplot(x="risk_category", y="performance_index", data=df, ax=axes[0])
axes[0].set_title("Performance Index vs Risk")

sns.boxplot(x="risk_category", y="consistency_score", data=df, ax=axes[1])
axes[1].set_title("Consistency Score vs Risk")

plt.show()
```

# Key Insights from EDA

Attendance and internal marks show strong separation between risk groups

Backlog count is a major indicator of academic risk

Engineered features (performance_index, consistency_score) improve class separability

Relationships are non-linear, justifying tree-based models
